# Fixing Duplicate Records

## Albums

```python
albums
```

```python
albums.shape[0] - albums["album_id"].nunique()
# there are 109 ids that are a copy of another id.
```

```python
# get the album_ids of the copies.
unique_ids = set()
duplicates = set()
for id in albums["album_id"]:
    if id in unique_ids:
        duplicates.add(id)
    unique_ids.add(id)
len(duplicates)
# there are 76 album_ids that copy another id at least once.
```

```python
filt = albums["album_id"].isin(duplicates)
filt.sum()
# there are 185 albums_ids that are similar to at least one of each other.
```

```python
all_duplicates = albums.loc[filt].sort_values("album_name")
all_duplicates.head(50)
```

### PROBLEM:
if two of the 20 artists show up on the same album, then that album will show up twice, but both copies will only have the artist_id of the artist that was listed first in the JSON response.

this problem is mainly due to albums that each artist "appears on" (meaning not their own album or song), as they work with similar artists that are also one of the 20 selected artists. 

### SOLUTIONS: 
either find the correct artist while making API requests, or remove all duplicates.

if all of the duplicates are removed, then some artist's will lose their "appears on" albums and collaborations.

### CHOSEN SOLUTION:
place the correct artist_id with each album, even if it was an album that they just "appeared on", because that is still technically one of their works, and should be included in the dataset.

#### DUPLICATES:
the only duplicates from "albums" that should be removed are records that have the same album_id, and the same artist_id.


```python
# find any albums that have the same artist_id listed twice
print(all_duplicates[["album_id", "artist_id"]])
all_duplicates[["album_id", "artist_id"]].duplicated().sum() # proving there are no album/artist duplicates.
```
